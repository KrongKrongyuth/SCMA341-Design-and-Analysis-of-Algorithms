# BFS

```python
def bfs(graph, start, end):
    
    foundNotHandled = [start]
    
    handled = []
    
    path = []
    
    rootOf = {}
    

    
    while len(foundNotHandled) > 0:
        print(handled)
        current = foundNotHandled.pop(0)
        print("found not handled", foundNotHandled, "current", current)
        
        
        if current == end:
            
            # trace back from end to start
            path = [current]
            
            
            while path[0] in rootOf:
                root = rootOf[path[0]]
                path = [root] + path
            return(path)
        
        elif current not in (handled):
            handled.append(current)
        
            foundNotHandled = foundNotHandled + graph[current]
           
            for node in graph[current]:
                if node not in rootOf:
                    rootOf[node] = current
            print("root", rootOf)
            
    return([])
    
    
    

```

```python
p = bfs(graph, "A", "E")
print(p)
```

# Dijkstra

```python
graph = [[0, 4, 0, 0, 0, 0, 0, 8, 0],
           [4, 0, 8, 0, 0, 0, 0, 11, 0],
           [0, 8, 0, 7, 0, 4, 0, 0, 2],
           [0, 0, 7, 0, 9, 14, 0, 0, 0],
           [0, 0, 0, 9, 0, 10, 0, 0, 0],
           [0, 0, 4, 14, 10, 0, 2, 0, 0],
           [0, 0, 0, 0, 0, 2, 0, 1, 6],
           [8, 11, 0, 0, 0, 0, 1, 0, 7],
           [0, 0, 2, 0, 0, 0, 6, 7, 0]
           ]

graph
```

```python
import math

def dijk(graph, start, end):
    dist = [math.inf] *len(graph)
    fromNode = [-1]*len(graph)

    
    dist[start] = 0
    
    handled = []
    
    handled.append(start)
    
    while end not in handled:
        
        # Get the most recent node
        lastNode = handled[-1]
        
        
        # Try every connection from the most recent node
        for n in range(len(graph[lastNode])):
            if (graph[lastNode][n] >0) and (n not in handled):
        
        
                # Update new distance from recently handled node
                newDist = dist[lastNode] + graph[lastNode][n]
                
                # Relax
                if newDist < dist[n]:
                    dist[n]  = newDist
                    fromNode[n] = lastNode
                    print("now relax distance of node", n,  "to be", newDist)
        
        
        print("current distance", dist)
        
        
        # Choose shortest unhandled to be next handle
        nextHandled = -1
        minDist = math.inf
        
        for m in range(len(graph)):
            if m not in handled:
                
                if dist[m] < minDist:
                    minDist = dist[m]
                    nextHandled = m
        handled.append(nextHandled)
        
        print("Handled",  handled, "\n\n")
        
    print("min distance", dist)
    print("Parents", fromNode)
    
    
    #generate path
    path = [end]
    while path[0] != start:
        path = [fromNode[path[0]]] + path
        
    print("path", path)
    
    
    
dijk(graph, 0, 5)
        
```
