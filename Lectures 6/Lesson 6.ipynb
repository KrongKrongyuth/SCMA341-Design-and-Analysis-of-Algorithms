# Tower of Hanoi

```python
import copy
def toh(poles, source, dest, spare, k): 
    
    # want to move k disks on top of source pole
    print("Want to move the top", k , "disks from", source, " to ", dest)
    
    # Base case
    if k == 1:
        #Move one disk
        d = poles[source].pop(0)    # Taking the first elements from source.
        ###################
        poles[dest] = [d] + poles[dest]
        
        # report when a disk actually move
        print("Now moving ",d," from pole", source, "to", dest)
        print(poles, "\n")
        return(poles)
    
    # Recursive case
    else:
        # Move top k-1 away
        #####################
        p1 = toh(poles, source, spare, dest, k-1)   # Same size as p3
        
        # Move the largest disk to the dest
        #####################
        p2 = toh(p1, source, dest, spare, 1)
        
        # Move top k-1 back on top of the largest
        #####################
        p3 = toh(p2, spare, dest, source, k-1)     # Same size as p1
        
        return p3

poles = [list(range(1,10)), [], []]

k = len(poles[0])
toh(poles, 0, 2, 1, k)
print("Slover finished!")
```

# Evaluate mathematical expression


```python
class Node:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data
```

```python
# Manual construction for expression tree of "((5-7)+(2-(4*3)))"

n9 = Node("3")
n8 = Node("4")
n7 = Node("*")
n7.left = n9
n7.right = n8
n6 = Node("2")
n5 = Node("-")
n5.left = n7
n5.right = n6
n4 = Node("7")
n3 = Node("5")
n2 = Node("-")
n2.left = n3 
n2.right = n4
n1 = Node("+")
n1.left = n5
n1.right = n2
```

```python
# Traverse the binary and print with indents
def printTree(n, k):
    
    if n.left == None and n.right == None:
        print("    "*k, n.data)
    else:
        printTree(n.left, k+1)
        print("    "*k, n.data)
        printTree(n.right, k+1)

printTree(n1, 0)
```

```python
# Evaluate expression tree
def evaluate(n):
    
    if n.data.isnumeric():
        return(int(n.data))
    else:
        
        # Evaluate left and right subtree
        l = evaluate(n.left)
        r = evaluate(n.right)
        
        # Do the middle operator with left and right outputs
        operation = n.data
        if operation == "+": return l + r
        elif operation == "-": return l - r
        elif operation == "*": return l * r
        elif operation == "/": return l / r
evaluate(n1)
```

# Recursive multiplication



```python
def mul(x,y):
    
    if y > x:
        x, y = y, x
        
        
    x = str(x)
    y = str(y)
    n1 = len(x)
    n2 = len(y)
    n = max(n1,n2)
    
    if n1 == 1 and n2 == 1:
        return(int(x) * int(y))
    
    else:
        a = x[:n1 - n//2]
        b = x[n1 - n//2:]
        c = y[:n2 - n//2]
        d = y[n2 - n//2:]

        if a == "":
            a = "0"
        if b == "":
            b = "0"
        if c == "":
            c = "0"
        if d == "":
            d = "0"
            
        #print(x,y)
        
        
        ac = mul(a,c)
        bd = mul(b,d)
        ad = mul(a,d)
        bc = mul(b,c)
        
        #print(ac, bd, ad, bc)
        s = ac*10**(2*(n//2)) + (ad+bc)*10**(n//2) + bd
        #print(s)
        return(s)
    
    

s = mul(5432354556854354344354384351238434387873543542541, 3563354354354354354354397852547874257845354464)
print(s)
```

```python
def karatsuba(x,y):
    
    if y > x:
        x, y = y, x
        
        
    x = str(x)
    y = str(y)
    n1 = len(x)
    n2 = len(y)
    n = max(n1,n2)
    
    if n1 == 1 and n2 == 1:
        return(int(x) * int(y))
    
    else:
        a = x[:n1 - n//2]
        b = x[n1 - n//2:]
        c = y[:n2 - n//2]
        d = y[n2 - n//2:]

        if a == "":
            a = "0"
        if b == "":
            b = "0"
        if c == "":
            c = "0"
        if d == "":
            d = "0"
            
        ac = karatsuba(a,c)
        bd = karatsuba(b,d)
        adbc = karatsuba(int(a)+int(b), int(c)+int(d)) - ac - bd
        #print(ac, bd, adbc)
        
        s = ac*10**(2*(n//2)) + adbc*10**(n//2) + bd
        #print(s)
        return(s)


s= karatsuba(5432354556854354344354384351238434387873543542541, 3563354354354354354354397852547874257845354464)
print(s)
```
