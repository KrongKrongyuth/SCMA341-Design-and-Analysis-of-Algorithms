Algorithm to check whether a sorted list has got a repeated element or not (return True or False)

```python
L  = [1,2,3,4,4,5,6,7,8,8,8,9,10,11,14,17,17,19]
#L = [1,2,3]
def check_repeat(L):
    i = 0
    while i < len(L)-1:
        if L[i] == L[i+1]:
            return(True, i)
        
        i+= 1
    return(False, i)
    
```

Algorithm to count how many pairs of repeated elements (return integer)

```python
def count_repeat(L):
    
    counter = 0
    loop_counter = 0
    for i in range(len(L)-1):
        if L[i] == L[i+1]:
            counter += 1
        loop_counter += 1
    return(counter, loop_counter)
count_repeat(L)
```

Generator for a randomized list containing sorted elements of length $k$

```python
import random
def random_generator(k):
    r = [random.randint(1, 100*k) for _ in range(k)] 
    r.sort()
    return(r)


numloop_check_repeat = []
numloop_count_repeat = []

for i in range(1000):
    
    L = random_generator(i)
    #print(L)
    c1 = check_repeat(L)[1]
    c2 = count_repeat(L)[1]
    numloop_check_repeat.append(c1)
    numloop_count_repeat.append(c2)
```

```python
import matplotlib.pyplot as plt


plt.plot(numloop_check_repeat, "r")
plt.plot(numloop_count_repeat, "g")
plt.show()
```

# Max of the list

```python
import math
def maxList(L):
    maxSoFar = -math.inf
    for i in range(len(L)):
        if L[i] > maxSoFar:
            maxSoFar = L[i]
        
    return(maxSoFar)

maxList([2,17, 45, 43, 14, 6, 18, 97, 24, 63, 57])
```

# Selection sort

```python
import math

def selectionSort(L):
    
    sorted_list = []
    for k in range(len(L)):
        minSoFar = math.inf
        pos = 0
        for i in range(len(L)):
            if L[i] < minSoFar:
                minSoFar = L[i]
                pos = i
        L.pop(pos)
        
        sorted_list.append(minSoFar)
        
    return(sorted_list)
    
selectionSort([45, 43, 24, 14, 6, 18, 97, 24, 63, 57])
```
